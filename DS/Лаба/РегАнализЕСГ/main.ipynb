```python
from os import listdir
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
str_ratings = [f"{letter}{sign}" for letter in ["D", "C", "B", "A"] for sign in ["-", "", "+"]]
values_rating = {str_ratings[i]: 0 + (1/(len(str_ratings)-1))*i for i in range(len(str_ratings))}
```

```python
def rating2num(str):
    # перобразуем категорию в численную оценку
    return values_rating[str]

print(values_rating)
```

```python
def data_process(df):
    
    # обработка части с оценками esg (обрезать - рейтинг - группировка)
    industry_data_esg = df.loc[df.question_type == "cat", "pillars"::]
    industry_data_esg.iloc[:, 1:] = industry_data_esg.iloc[:, 1:].applymap(rating2num)
    industry_data_esg = industry_data_esg.groupby('pillars').mean().reset_index()

    # обработка части с фин оценками (обрезать - заменить столбец - удалить лишнее)
    industry_data_fin = df.loc[~df.pillars.isin(["E", "S", "G"]) & ~df.questions.isin([np.NAN]), :].copy()
    industry_data_fin["pillars"] = industry_data_fin["questions"]
    industry_data_fin = industry_data_fin.drop(["question_type", "questions", "Unnamed: 0"], axis = 1)

    # соединяем датафреймы 
    res = pd.concat([industry_data_esg, industry_data_fin], ignore_index=True)

    # переворачиваем датасет
    res = res.set_index("pillars").T   

    return res
```

```python
data_files = listdir("./data")

for industry_file in data_files[:5]:

    industry_data = pd.read_excel(f"data/{industry_file}")
    
    data = data_process(industry_data)

    # print(industry_file, "\n")
    # display(data)

data 
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Пример данных
data = {'X1': np.array([1, 2, 3, 4, 5]),
        'X2': np.array([2, 3, 4, 5, 6]),
        'X3': np.array([3, 4, 5, 6, 7]),
        'y': np.array([5, 8, 10, 12, 14])}

df = pd.DataFrame(data)

# Добавляем столбец с константой для коэффициента сдвига
df['const'] = 1

# Создание и обучение модели OLS
model = sm.OLS(df['y'], df[['const', 'X1', 'X2', 'X3']])
results = model.fit()

# Вывод результатов
print(results.summary())

# График регрессии
fig, ax = plt.subplots()
ax.scatter(df['X1'], df['y'], color='red', label='X1')
ax.scatter(df['X2'], df['y'], color='green', label='X2')
ax.scatter(df['X3'], df['y'], color='blue', label='X3')

ax.plot(df['X1'], results.predict(df[['const', 'X1', 'X2', 'X3']]), color='black', linewidth=3, label='Regression Line')
ax.legend()

plt.title('OLS Regression with 3 Independent Variables')
plt.xlabel('X1, X2, X3')
plt.ylabel('y')
plt.show()
```
